```python
import konlpy 
```

```python
txt = open('C:\\Users\\82108\\Desktop\\Ybigta\\02_04 wordcloud konlpy\\세션자료\\실습\\data\\위대한 유산.txt').read()
txt
```

```python
from collections import Counter
```

```python
okt=konlpy.tag.Okt()
```

```python
noun = okt.nouns(txt)
for i,v in enumerate(noun):
    if len(v) < 2:
        noun.pop(i)
        
count = Counter(noun)
noun_list = count.most_common(100)
noun_list
```

```python
del noun_list[0:9]
```

```python
txt_wc = WordCloud(font_path='C:\\Users\\82108\\Desktop\\Ybigta\\02_04 wordcloud konlpy\\세션자료\\실습\\data\\NanumSquareB.ttf',
                   background_color = "white",
                   collocations = False,
                   mask = mask_array,
                   stopwords = stopwords,
                   max_words=300,
                   max_font_size=200) # 앞에서 np.array로 만든 mask 지정

txt_wc.generate_from_frequencies(dict(noun_list))
txt_wc.to_image()
```

```python
txt_wc.to_file("위대한유산.png")
```
